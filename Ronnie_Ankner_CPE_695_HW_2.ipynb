```python
import numpy as np
import matplotlib.pyplot as plt
from time import time

np.set_printoptions(formatter={'float': lambda x: "{0:0.3f}".format(x)})

def calculate_gradient(w, Y, X):
    grad_w = np.array([0]*len(w))
    
    #Check to make sure X is correct orientation
    #print("Before: ", X.shape)
    if (len(X.shape) > 1):
        if (X.shape[1] != len(w)):
            #if not, try flipping
            print("Incorrect dimensioning, trying transpose")
            X = X.T

        #if still not equal, return with nothing because samples are formatted improperly
        if (X.shape[1] != len(w)):
            print("Error, samples are not the same dimension as w")
            return grad_w
    else:
        if (X.shape[0] != len(w)):
            print("Error, samples are not the same dimension as w")
            return grad_w

    m = X.shape[0]
    #print("Number of Samples: ", m)
    #print("Dimension of samples: ", X.shape[1])
    
    grad_w = (2/m)*np.dot(X.T, (np.dot(X,w)-Y))
    
    return grad_w
#============================ Create Data Set ==========================================

data_file = open('data2.txt','r')
data = data_file.read()
pairs = data.split('\n')

x = []
y = []
cnt = 0
for pair in pairs:
    data_point = pair.split(',')
    if (len(data_point) == 2):
        cnt += 1
        #print("Sample %d: "%(cnt), pair)
        x.append(float(data_point[0]))
        y.append(float(data_point[1]))

samples = []
outputs = []
power2fit = 2
for val in x:
    samp = []
    for i in range(0,power2fit):
        samp.append(val**i)
    samples.append(samp)

outputs = np.array(y)
samples = np.array(samples)


X = samples
inv = np.linalg.inv(np.dot(X.T, X))
P = np.dot(inv, X.T)
truth = np.dot(P, y)
print("Closed form Regression Solution: ", truth)

true = np.dot(X, truth)

plt.plot(x, outputs, 'bo', x, true, 'g')
plt.title("Sample Data Set")
plt.show()
```

```python
# ==================== Batch Learning ===========================
w = np.array([0]*samples.shape[1])

cnt = 0
cnt_list = []
diff = 100000
watchdog = 20000
coeff_list = []
#Training Parameters
nu = 0.01188
tol = 0.005

start = time()
while diff > tol:
    coeff_list.append(w.tolist())
    cnt_list.append(cnt)
    
    cnt += 1
    #print("Step #%d"%(cnt))
    grad = calculate_gradient(w, outputs, samples)
    diff = abs(np.sum(np.dot(grad, grad)))
    #print(grad)
    #print(nu*grad)
    
    w = w - nu*grad
    #print(w)
    
    if cnt == watchdog:
        print("MINIMUM NOT FOUND! Exited after %d Steps"%(cnt), end = ' ')
        break
        
total_time = time()-start

if (cnt < watchdog):
    print("System converged in %d steps"%(cnt), end = ' ')
print("After %.4f seconds"%(total_time))
print("Final w:")
print(w, '\n')
print("Difference from Truth")
print(abs(w-truth), '\n')

coeff_list = np.array(coeff_list)

legend_names = []

for i in range(0, coeff_list.shape[1]):
    plt.plot(cnt_list, coeff_list[:,i])
    legend_names.append(str(i+1))
    
    
plt.title("Coefficents vs Step")
plt.legend(legend_names)
plt.show()

y_hat = np.dot(w, samples.T)
true = np.dot(truth, samples.T)

plt.plot(x, outputs, 'bo', x, y_hat, 'r*', x,true, 'g')
plt.title("Y-hat vs Sample Values")
plt.legend(["Data", "Y-hat", "Truth"])
plt.show()
```

```python
# ====================== Stochastic Descent ==============================
w = np.array([0]*samples.shape[1])

cnt = 0
cnt_list = []
diff = 100000
watchdog = 10000
coeff_list = []
#Training Parameters
nu = 0.001

num_samples = samples.shape[0]

start = time()
while cnt < watchdog:
    coeff_list.append(w.tolist())
    cnt_list.append(cnt)
    
    cnt += 1
    #print("Step #%d"%(cnt))
    rand_samp = int(np.random.uniform(0,num_samples,1)[0])
    X = samples[rand_samp, :]
    Y = outputs[rand_samp]
    grad = calculate_gradient(w, Y, X)
    diff = abs(np.sum(np.dot(grad, grad)))
    #print(grad)
    #print(nu*grad)
    
    w = w - nu*grad
    #print(w)
        
total_time = time()-start

print("System converged in %d steps"%(cnt), end = ' ')
print("After %.4f seconds"%(total_time))
print("Final w:")
print(w, '\n')
print("Difference from Truth")
print(abs(w-truth), '\n')
coeff_list = np.array(coeff_list)

legend_names = []

for i in range(0, coeff_list.shape[1]):
    plt.plot(cnt_list, coeff_list[:,i])
    legend_names.append(str(i+1))
    
    
plt.title("Coefficents vs Step")
plt.legend(legend_names)
plt.show()

y_hat = np.dot(w, samples.T)
true = np.dot(truth, samples.T)

plt.plot(x, outputs, 'bo', x, y_hat, 'r*', x,true, 'g')
plt.title("Y-hat vs Sample Values")
plt.legend(["Data", "Y-hat", "Truth"])
plt.show()
```

# Determining the Best Learning Rate

## Batch Learning Rate

```python
import math
#============= Selecting Best Learning Rate ==================
#================== Batch ==============================
nu_list = [0.00001]
for i in range(1,100):
    nu_list.append(nu_list[i-1]*1.08)
    
print("Range Nu_list: [%f, %f]"%(nu_list[0],nu_list[99]))

efficiency = [] #(sum square error)*(steps_to_converge) -> best will be lowest value
sse_s = []
converge = []

#for nu in nu_list:
start = time()
for nu in nu_list:
    w = np.array([0]*samples.shape[1])

    cnt = 0
    diff = 100000
    watchdog = 20000
    #Training Parameters
    tol = 0.005

    while diff > tol:
        cnt += 1
        #print("Step #%d"%(cnt))
        grad = calculate_gradient(w, outputs, samples)
        diff = abs(np.sum(np.dot(grad, grad)))
        #print(grad)
        #print(nu*grad)

        w = w - nu*grad
        #print(w)

        if cnt == watchdog:
            #print("MINIMUM NOT FOUND! Exited after %d Steps"%(cnt), end = ' ')
            break

    err = w-truth
    sse = np.dot(err, err)
    
    if (not(sse > 0 and sse < 30)):
        #penalize non-convergence and account for w going to inf before watchdog
        sse = 30
        cnt = watchdog
        
        
    converge.append(cnt)
    sse_s.append(sse)
    efficiency.append(sse*cnt)
    
run_time = time()-start
print("Run Completed in %.2f s"%(run_time))

min_eff = 10000000000
ind = 0
cnt = 0
for eff in efficiency:
    if eff < min_eff:
        min_eff = eff
        ind = cnt
    cnt += 1
nu_best = [nu_list[ind]]
eff_best = [min_eff]

plt.plot(nu_list, efficiency, 'b', nu_best, eff_best, 'ro')
plt.title("Efficiency vs Learning Rate")
plt.show()

print("Best Learning Rate for Batch and this data set is: %.5f"%(nu_list[ind]))
print("\t Steps to Converge: ", converge[ind])
print("\t Sum Square Error : %.5f"%(sse_s[ind]))
```
